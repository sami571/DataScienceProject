```python
from pytrends.request import TrendReq
import pandas as pd
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["Gat","STAR","COMAR","ASTREE","MAGHREBIA","CARTE","LLOYD TUNISIEN","ASSURANCES SALIM","BIAT-ASSURANCES","ZITOUNA TAKAFUL","EL AMANA TAKAFUL","ATTAKAFULIA","AMI"]
```

```python
df=pd.DataFrame()
```

```python
for kw in kw_list:
    pytrends.build_payload([kw], cat=0, timeframe='today 5-y', geo='TN', gprop='')
    interest=pytrends.interest_by_region(resolution='kw', inc_low_vol=True, inc_geo_code=False)
    df=pd.concat([df, interest], axis=1)
```

```python
df=df.drop(['G', 'a', 't', 'S', 'T', 'A', 'R', 'C', 'O', 'M', 'A', 'R', 'G', 'a',
       't', 'S', 'T', 'A', 'R', 'C', 'O', 'M', 'A', 'R', 'G', 'a', 't', 'S',
       'T', 'A', 'R', 'C', 'O', 'M', 'A', 'R', 'G', 'a', 't', 'S', 'T', 'A',
       'R', 'C', 'O', 'M', 'A', 'R'],axis=1)
df.to_excel("Compagnies_par_regions.xlsx")
```
